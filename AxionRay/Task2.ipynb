### **Univariate Analysis - Continuous Variables**

```python
import scipy.stats as stats

def plot_continuous_data_analysis(df, column_name):

    # Create a 1x3 subplot layout for the histogram, boxplot, and Q-Q plot
    fig, axs = plt.subplots(1, 3, figsize=(20, 6), squeeze=False)


    plt.suptitle(f"Analysis of {column_name}", fontsize=15)


    # Histogram with KDE and mean, median, mode lines
    sns.histplot(df[column_name], kde=True, edgecolor='green', ax=axs[0, 0], bins=20)
    axs[0, 0].axvline(df[column_name].mean(), ls='--', color='red', lw=2.5, label='Mean')
    axs[0, 0].axvline(df[column_name].median(), ls='--', color='blue', lw=2.5, label='Median')
    axs[0, 0].axvline(df[column_name].mode()[0], ls='--', color='orange', lw=2.5, label='Mode')
    axs[0, 0].set_title("Histogram with KDE")
    axs[0, 0].legend()


    # Boxplot
    sns.boxplot(x=df[column_name], showmeans=True, ax=axs[0, 1])
    axs[0, 1].set_title("Boxplot")


    # Q-Q plot to check the normality of the data
    stats.probplot(df[column_name], dist="norm", plot=axs[0, 2])
    axs[0, 2].set_title("Q-Q Plot")


    plt.tight_layout()
    plt.show()


for i in ['REPAIR_AGE', 'KM',  'REPORTING_COST', 'TOTALCOST', 'LBRCOST']:
    plot_continuous_data_analysis(df1, i)
```

## **Outlier Treatment**

```python
def outlierAnalysis(x):
    Q3, Q1 = np.percentile(df[x], [75 ,25])
    IQR = Q3 - Q1
    maxExcludingOutlier = Q3 + 1.5 * IQR
    minExcludingOutlier = Q1 - 1.5 * IQR
    mean = round(df[x].mean(),2)
    median = round(df[x].median(),2)
    mode = round(df[x].mode(), 2)
    print("For the given sample " +x +" Analysis is as Follows: ")
    print("Q1: ", Q1)
    print("Q3: ", Q3)
    print("Mean: ", round(df[x].mean(),2))
    print("Median: ", df[x].median())
    print("Mode: " , round(df[x].mode(), 2))
    print("IQR: " , IQR)
    print("Maximum " + x +" Excluding Outlier: " , maxExcludingOutlier.round(3))
    print("Minimum " + x + " Purchase Excluding Outlier: " , minExcludingOutlier.round(3))


for x in ['REPAIR_AGE', 'KM',  'REPORTING_COST', 'TOTALCOST', 'LBRCOST']:
    print("Statistical information for ", x.upper())
    print(df1[x].describe().round(2))
    print("--"*40)
    print("Outlier Analysis for ", x.upper())
    print(outlierAnalysis(x))
    print("__"*60)
    print()
```

```python
# Outlier removal

import pandas as pd

def remove_outliers_iqr(df):

    for column in ['REPAIR_AGE', 'REPORTING_COST', 'TOTALCOST', 'LBRCOST']:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)

        # Calculate IQR
        IQR = Q3 - Q1

        # Define the bounds (1.5 * IQR rule)
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter the data by removing rows outside of the bounds
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df


df1 = remove_outliers_iqr(df1)
```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to preprocess the text (lowercase, remove punctuation)
def preprocess_text(text):
    doc = nlp(text.lower())  # Convert text to lowercase
    return " ".join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])

# Function to categorize the components and generate tags for 'CAUSAL_PART_NM'
def categorize_causal_part(part):
    category = []

    # Categorizing based on part type (e.g., wheel assembly, module, cover)
    if 'WHEEL ASM' in part or 'WHEEL,STRG' in part:
        category.append('Wheel Assembly')
    if 'MODULE' in part:
        category.append('Module')
    if 'HARNESS' in part:
        category.append('Harness')
    if 'COVER' in part:
        category.append('Cover')
    if 'APPLIQUE' in part:
        category.append('Applique')

    # Material or Finish (color, texture)
    if 'JET BLACK' in part:
        category.append('Color: Jet Black')
    if 'BACKEN BLACKK' in part:
        category.append('Color: Backen Black')
    if 'VERY DARK' in part:
        category.append('Color: Very Dark')
    if 'DARK TITANIU' in part:
        category.append('Color: Dark Titanium')
    if 'BLACK' in part:
        category.append('Color: Black')
    if 'LINEAR GALAXD' in part:
        category.append('Finish: Linear Galaxy')
    if 'VULCAN' in part:
        category.append('Finish: Vulcan')
    if 'HI GLOSS' in part:
        category.append('Finish: High Gloss')
    if 'SYNTHESIS' in part:
        category.append('Finish: Synthesis')
    if 'FAWN' in part:
        category.append('Color: Fawn')
    if 'GALVANIE' in part:
        category.append('Finish: Galvanized')

    if not category:
        category.append('Other')

    # Return as a comma-separated string instead of a list
    return ', '.join(category)


# Apply the categorization function to the 'CAUSAL_PART_NM' column and create a new 'casualpart_tag' column
df1['casualpart_tag'] = df1['CAUSAL_PART_NM'].apply(categorize_causal_part)
```

```python
import re

# Function to preprocess the text (lowercase, remove punctuation)
def preprocess_text(text):
    # Remove special characters and convert to lowercase
    text = re.sub(r'[^\w\s]', '', text.lower())
    return text

# Function to categorize and generate tags for 'CUSTOMER_VERBATIM'
def generate_tags_from_verbatim(text):
    tags = []

    # Preprocess the text
    text = preprocess_text(text)

    # Categorize based on keywords
    if 'engine' in text:
        tags.append('Engine Issue')
    if 'transmission' in text or 'slipping' in text:
        tags.append('Transmission Issue')
    if 'battery' in text or 'charging' in text:
        tags.append('Battery Issue')
    if 'electrical' in text or 'lights' in text:
        tags.append('Electrical Issue')
    if 'noise' in text:
        tags.append('Noise Issue')
    if 'brake' in text:
        tags.append('Brake Issue')
    if 'overheats' in text or 'temperature' in text:
        tags.append('Overheating Issue')
    if 'start' in text or 'won\'t start' in text:
        tags.append('Starting Issue')
    if 'power' in text or 'accelerating' in text:
        tags.append('Power Loss Issue')

    # If no specific tag matched, add 'Other'
    if not tags:
        tags.append('Other')

    # Return as a comma-separated string instead of a list
    return ', '.join(tags)

# Apply the function to the 'CUSTOMER_VERBATIM' column and create a new 'customer_tag' column
df1['customer_tag'] = df1['CUSTOMER_VERBATIM'].apply(generate_tags_from_verbatim)
```
